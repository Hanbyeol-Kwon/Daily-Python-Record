```python
import sqlite3


# sqlite 데이터베이스 연결하기 --- (※1)
dbpath = "test.sqlite"
conn = sqlite3.connect(dbpath)


# 테이블을 생성하고 데이터 넣기 --- (※2)
cur = conn.cursor()
cur.executescript()
conn.commit()


# 데이터를 추출하여 출력하기 --- (※3)
cur = conn.cursor()
cur.execute("SELECT item_id, name, price FROM items")
item_list = cur.fetchall()


# 출력하기
for it in item_list:
    print(it)
```
