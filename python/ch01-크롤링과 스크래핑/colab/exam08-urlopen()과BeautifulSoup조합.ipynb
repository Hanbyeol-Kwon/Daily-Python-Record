```python
import urllib.request as req #urllib.request의 별칭 req 사용
from bs4 import BeautifulSoup

url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
res = req.urlopen(url)
# 웹 브라우저에서 열거나 urllib.request로 가져온 res 객체의 내용을 print(res.read().decode('utf-8'))와 같이 출력해보면 원본 데이터를 볼 수 있음


soup = BeautifulSoup(res, "lxml")
#wf 태그를 찾을 수 없어서 오류 발생
title = soup.find("title").string
print(title)

# Find the first 'item' tag
item = soup.find("item")

if item:
    # Find the 'description' tag within the item
    description_tag = item.find("description")
    if description_tag and description_tag.string:
        # Parse the content of the description tag as a new BeautifulSoup object
        description_soup = BeautifulSoup(description_tag.string, "lxml")
        # Now find the 'wf' tag within this new soup object
        wf_tag = description_soup.find("wf")
        if wf_tag and wf_tag.string:
            wf = wf_tag.string
            print(wf)
        else:
            wf = "날씨 정보를 찾을 수 없습니다."
            print(wf)
    else:
        wf = "예보 상세 정보를 찾을 수 없습니다."
        print(wf)
else:
    wf = "예보 항목을 찾을 수 없습니다."
    print(wf)
```
