```python
from bs4 import BeautifulSoup

html = """
    <html>
        <body>
            <h1 id="title">스크래핑이란?</h1>
            <p id="body">웹 페이지를 분석하는 것</p>
            <p>원하는 부분을 추출하는 것</p>
        </body>
    </html>
"""

soup = BeautifulSoup(html, 'html.parser')
# Beautifulsoup soup = new Beautifulsoup(html,'html.parser')

title = soup.find(id="title") #5번 예제와 같은 문법으로 soup. 으로 사용한다.
body = soup.find(id="body") # id명을 사용하는 건 주로 find()- 하나만 추출, class명을 주로 이용하는 건 select()- 여러 개 추출
# soup.select('.my-class')는 my-class 클래스명을 가진 요소의 텍스를 리스트형으로 추출

print("#title=" + title.string)
print("#body=" + body.string)
```
