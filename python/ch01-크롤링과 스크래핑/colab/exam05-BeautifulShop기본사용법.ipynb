```python
from bs4 import BeautifulSoup
# BeautifulSoup는 데이터를 쉽게 추출할 수 잇도록 돕는 도구
# 설치: 명렁프롬프트(cmd or terminal)에서 'pip install beautifulsoup4' 입력 후 [Enter] 눌러서 설치
# find()로 특정 조건을 만족하는 요소를 찾음
# select('클래스선택자')로 클래스 선택자에 해당하는 요소를 리스트형[]으로 가져옴
# get_text()로 요소안에 텍스트만 추출
# 모듈을 이용한 원하는 부분 데이터 추출


html = """ 데이터를 가져왔다고 가정하 실습
<html>
    <body>
        <h1>스크래핑이란?</h1>
        <p>웹 페이지를 분석하는 것</p>
        <p>원하는 부분을 추출하는 것</p>
    </body>
</html>
"""

soup = BeautifulSoup(html, 'html.parser') #객체 생성
#html.parser는 html을 분석하는 분석기(parser)이다.

h1 = soup.html.body.h1
p1 = soup.html.body.p
p2 = p1.next_sibling.next_sibling
# 첫번째, next_sibling을 줄바꿈을 추출한다.
# 두번째, next_sibling을 p 요소의 텍스트를 추출한다.

print("h1 = " + h1.string)
print("p  = " + p1.string)
print("p  = " + p2.string)
```
