```python
import urllib.request
import urllib.parse #urlencode() 사용을 위해 임포트

#기상청의 데이터를 가져와서 날씨에 대한 데이터를 크롤링하는 예제
API = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp" #상수라서 대문자로 작성

values = { #딕셔너리형
  'stnId': '182'} #109:서울경기 105:강원도 182:제주도

params = urllib.parse.urlencode(values) #stnld=108 url형식으로 바꿈

url = API + "?" + params
# url 변수에 "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?"
print("url=", url)

data = urllib.request.urlopen(url).read()
text = data.decode("utf-8")

print(text)
```
