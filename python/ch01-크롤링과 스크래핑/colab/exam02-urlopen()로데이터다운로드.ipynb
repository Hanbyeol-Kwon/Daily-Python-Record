```python
import urllib.request

url = "http://uta.pw/shodou/img/28/214.png"
savename = "test.png"

mem = urllib.request.urlopen(url).read()
#url을 메모리상에 저장하고 링크를 연다

#컴퓨터 언어(이진수)로 작성하고, 바이너리 개념으로 쓰겠다는 모드(설정)을 f.약칭으로 저장
# mode: w(쓰기) r(읽기) wb(바이너리, 이진수)
# open으로 열었으면 close로 꼭 닫아줘야 하는데 'with'는 f.close 역할을 자동 수행
with open(savename, mode="wb") as f:
    f.write(mem)

print("저장되었습니다...!")
```
