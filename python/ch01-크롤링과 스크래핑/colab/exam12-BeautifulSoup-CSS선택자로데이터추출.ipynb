```python
from bs4 import BeautifulSoup

fp = open("fruits-vegetables.html", encoding="utf-8")
soup = BeautifulSoup(fp, "html.parser")

print(soup.select_one("li:first-child").string) # 사과
print(soup.select_one("#fr-list > li:nth-child(4)").string) # 오렌지
print(soup.select("#ve-list > li[data-lo='us']")[1].string) #[파프리카, (아보카도)]
print(soup.select("#ve-list > li.black")[1].string) #[가지, (아보카도)]

cond = {"data-lo":"ko", "class":"black"} # 딕셔너리형/가지
print(soup.find("li", cond).string) # 가지
# find()는 여러 개의 조건을 한번에 지정
# li 요소를 찾음. data-lo 속성값이 ko인 경우, class 속성값이 black인 요소의 데이터를 추출
print(soup.find(id="ve-list").find("li", cond).string)
```
