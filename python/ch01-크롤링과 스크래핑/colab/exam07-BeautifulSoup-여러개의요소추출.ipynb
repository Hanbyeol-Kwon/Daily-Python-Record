```python
from bs4 import BeautifulSoup

# lxml 파서 설치 (만약 설치되어 있지 않다면)
!pip install lxml
# !pip install html.parser --- 가능한가 시도했지만, 설치는 되지만 BeautifulSoup에서 적합한 트리 빌더를 찾지 못했기 때문에 같은 오류 발생
# 그럼에도 불구하고 Python에 내장된 html.parser 모듈을 사용하고 싶다면? 순수 파이썬 파서를 사용해야 한다면, html5lib가 훌륭한 대안
# 아마 수업에서 사용한 파이썬 버전3.10과 내 노트북3.14의 파이썬 버전이 다르기 때문이라고 유추

html = """
    <html>
        <body>
            <ul>
                <li><a href="http://www.naver.com">naver</a></li>
                <li><a href="http://www.daum.net">daum</a></li>
            </ul>
        </body>
    </html>
"""

soup = BeautifulSoup(html, 'html5lib') # 'html.parser' 대신 'lxml' 파서 사용

links = soup.find_all("a") # 여러 요소를 리스트형으로 추출

for a in links: # 리스트형, range(): 반복 횟수 지정
    href = a.attrs['href'] # 링크의 href 속성 값은 attrs 속성에서 추출
    text = a.string # a 요소가 저장한 텍스트를 String형으로 추출
    print(text, ">", href)
```
