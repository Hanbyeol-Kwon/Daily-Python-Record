```python
from bs4 import BeautifulSoup
import urllib.request as req

url = "https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC"

# Add a User-Agent header to mimic a web browser
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
request = req.Request(url, headers=headers)

res = req.urlopen(request)
soup = BeautifulSoup(res, "html.parser")

# New selection logic
a_list = []
content_area = soup.find(id="mw-content-text")

if content_area:
    # Find the h2 that contains the text '작품 목록'
    works_sections = ['시집', '시'] # Target these sections based on debug output

    for h_tag in content_area.find_all(['h2', 'h3']): # Check both h2 and h3, to be robust
        h_text = h_tag.get_text().strip()
        if h_text in works_sections:
            works_list_ul = h_tag.find_next_sibling("ul")
            if works_list_ul:
                a_list.extend(works_list_ul.select("li a"))

print(f"Length of a_list: {len(a_list)}")

# Print a portion of the soup to inspect the HTML structure if a_list is empty
if not a_list:
    print("a_list is empty. Printing a portion of the HTML to debug:")
    content_div = soup.select_one("#mw-content-text > div")
    if content_div:
        print(content_div.prettify()[:1000])
    else:
        print("Could not find #mw-content-text > div")

# select()는 여러 요소를 얻어서 리스트형으로 저장

for a in a_list:
    name = a.string
    print("-", name)
```
