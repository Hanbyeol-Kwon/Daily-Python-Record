```python
from bs4 import BeautifulSoup # html 파일을 분석할 때 사용
from urllib.request import * # 인터넷에서 데이터 다운로드 받을 때
from urllib.parse import * # url 분석할 때
from os import makedirs # 폴더 생성 시
import os.path, time, re # 경로(os.path), time:슬립(wait), re:정규표현식

# 재귀적으로 html 파일을 다운로드
# 프로그래밍 기법 중 하나로 함수 내부에서 함수 자신을 호출하는 것
# 이런 기법을 이용해서 링크에 연결된 모든 html 파일 다운로드
# 1. html 분석 2. 링크 추출 3. 링크에 해당하는 html 파일 다운로드

proc_files = {}

def enum_links(html, base):
    soup = BeautifulSoup(html, "html.parser")
    links = soup.select("link[rel='stylesheet']")
    links += soup.select("a[href]")

    result = []

    for a in links:
        href = a.attrs['href']
        url = urljoin(base, href)
        result.append(url)

    return result




def download_file(url):
    o = urlparse(url)
    savepath = "./" + o.netloc + o.path

    if re.search(r"/$", savepath):
        savepath += "index.html"

    savedir = os.path.dirname(savepath)

    if os.path.exists(savepath): return savepath

    if not os.path.exists(savedir):
        print("mkdir=", savedir)
        makedirs(savedir)

    try:
        print("download=", url)
        urlretrieve(url, savepath)
        time.sleep(1)
        return savepath
    except:
        print("다운 실패: ", url)
        return None




def analyze_html(url, root_url):
    savepath = download_file(url)

    if savepath is None: return
    if savepath in proc_files: return

    proc_files[savepath] = True

    print("analyze_html=", url)

    html = open(savepath, "r", encoding="utf-8").read()
    links = enum_links(html, url)

    for link_url in links:
        if link_url.find(root_url) != 0:
            if not re.search(r".css$", link_url):
                continue

        if re.search(r".(html|htm)$", link_url):
            analyze_html(link_url, root_url)
            continue

        download_file(link_url)



if __name__ == "__main__":
    url = "https://docs.python.org/3.8/library/"
    analyze_html(url, url)
```
