### 예제 사이트 크롤링 (크롤링 코드 패턴)

```python
import requests
from bs4 import BeautifulSoup

url = 'https://davelee-fun.github.io/'
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('')
for item in items:
    print (item.get_text())
```

### 예제 사이트 크롤링 연습1
- Blog 와 About 메뉴 리스트 가져와 출력하기

```python
import requests
from bs4 import BeautifulSoup

url = 'https://davelee-fun.github.io/'
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')

menus = soup.select('ul.navbar-nav.ml-auto li.nav-item')
for menu in menus:
    print(menu.get_text().strip())
```

### 예제 사이트 크롤링 연습2

### 예제 사이트 크롤링 연습3
- '선물하기 좋은 선물을 고르고 골라서 소개하는 블로그입니다.' 서브타이틀 가져와 출력하기

```python
import requests
from bs4 import BeautifulSoup

url = 'https://davelee-fun.github.io/'
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')

head = soup.find('div', class_='mainheading')
subtit = head.find('p', class_='lead')
print(subtit.get_text().strip())
```

### 예제 사이트 크롤링 연습4
- Featured 에 있는 상품명 가져와 출력하기

```python
import requests
from bs4 import BeautifulSoup

url = 'https://davelee-fun.github.io/'
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')

features = soup.select('section.featured-posts div.card-body h4.card-text')
for feature in features:
    print(feature.get_text().split(',')[0].strip())
```

### 예제 사이트 크롤링 연습5
- All Stories 에 있는 상품명 가져와 출력하기

```python
import requests
from bs4 import BeautifulSoup

url = 'https://davelee-fun.github.io/'
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')

stories = soup.select('section.recent-posts div.card-body h4.card-text')
for story in stories:
    print(story.get_text().split(',')[0].strip())
```

### 예제 사이트 크롤링 연습6
- 'Never miss a story from us, subscribe to our newsletter' 하단부 메세지 가져와 출력하기

```python
import requests
from bs4 import BeautifulSoup

url = 'https://davelee-fun.github.io/'
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')

ftmsg = soup.select_one('div.container.text-center > span')
print(ftmsg.get_text().strip())
```
