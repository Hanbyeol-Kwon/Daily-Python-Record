### 엑셀 파일 만들기
- 함수로 작성해보면서, 함수 작성법과 활용법 이해하기

````python
import openpyxl
def write_excel_template(filename, sheetname, listdata):
    excel_file = openpyxl.Workbook() #파일 생성
    excel_sheet = excel_file.active
    excel_sheet.column_dimensions['A'].width=100 #열 너비 지정
    excel_sheet.column_dimensions['B'].width=50
    if sheetname !='':
        excel_sheet.title=sheetname
    for item in listdata:
        excel_sheet.append(item) #가져온 데이터가 끝날 때까지 행 단위로 추가
    excel_file.save(filename)
    excel_file.close #불필요한 프로세스를 없애기 위해 이용 후 꼭 닫기
````

### 크롤링해서 엑셀 파일까지 만들기
- 리스트 안에 리스트(각 행을 나타냄)가 들어가야 함

````python
import requests
from bs4 import BeautifulSoup

goods_lists = list() #첫번째 칼럼, 두번째 칼럼에서 가져온 데이터를 넣을 리스트 생성

for pnum in range(5):
    if pnum == 0: res = requests.get('https://davelee-fun.github.io/')
    else : res = requests.get('https://davelee-fun.github.io/page'+str(pnum+1))
    soup = BeautifulSoup(res.content, 'html.parser')

    data = soup.select('div.card')
    for item in data:
        goods_name = item.select_one('div.card-body h4.card-text')
        goods_podate = item.select_one('div.wrapfooter span.post-date')
        goods_info = [goods_name.get_text().strip(), goods_podate.get_text().strip()]
        goods_lists.append(goods_info)
````

````python
write_excel_template('products.xlsx', 'goods_info', goods_lists)
````
