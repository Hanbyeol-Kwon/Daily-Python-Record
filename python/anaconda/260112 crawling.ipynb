```python
from bs4 import BeautifulSoup
html = '''<html>
            <body>
                <h1 id='title'>[1]크롤링이란?</h1>
                <p class='cssstyle'>웹페이지에서 필요한 데이터를 추출하는 것</p>
                <p id='body' align='center'>파이썬을 중심으로 다양한 웹크롤링 기술 발달</p>
            </body>
        </html>'''
soup = BeautifulSoup(html,'html.parser')
# find_all() 함수 사용하기
datas = soup.find_all('p')
for data in datas:
    print(data.get_text())
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://davelee-fun.github.io/blog/crawl_test')
soup = BeautifulSoup(res.content, 'html.parser')
part = soup.find('ul',id='dev_course_list')
titles = part.find_all('li','course')

for index, title in enumerate(titles): # enumerate 넘버링 하는 함수
    print(str(index+1)+'.', title.get_text().split('-')[-1].split('[')[0].strip())
```

```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://davelee-fun.github.io/blog/crawl_html_css.html')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('tr')
for item in items:
    columns = item.select('td')
    row_str = ''
    for column in columns:
        row_str += ', '+column.get_text()
    print(row_str[2:])
```
